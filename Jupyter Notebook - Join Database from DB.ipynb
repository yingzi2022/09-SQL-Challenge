```python
# Query All Records in the the Table
deptmanager_data = pd.read_sql_query('select * from dept_manager', con=engine)
deptmanager_data
```

```python
# Query All Records in the the Table
departments_data = pd.read_sql_query('select * from departments', con=engine)
departments_data
```

```python
# Merge 1: data table employees and salaries

employees_salaries = pd.merge(employees_data, salaries_data, on="emp_no", how="inner")
employees_salaries
```

```python
employees_salaries.rename(columns = {'emp_title_id':'title_id'}, inplace = True)
employees_salaries
```

```python
# Merge 2: data table employees and salaries with title
employees_salaries_titles = pd.merge(employees_salaries, titles_data, on="title_id", how="left")
employees_salaries_titles
```

```python
#Add new column 1: dept_emp 
deptemp_data['employee type'] = 'employee'
deptemp_data
```

```python
#Add new column 2: dept_emp to departments
deptmanager_data['employee type'] = 'manager'
deptmanager_data
```

```python
combine_deptemp = pd.merge(deptemp_data, deptmanager_data, on="emp_no", how="left")
combine_deptemp
```

```python
# combine 1: data table dept_emp to departments
combine_dept = pd.merge(deptemp_data, deptmanager_data, on="emp_no", how="left", suffixes=('deptemp_data', 'deptmanager_data'))
combine_dept
```
